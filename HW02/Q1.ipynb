```python
import numpy as np
from matplotlib import pyplot as plt
import random
from scipy.stats import norm
import pandas as pd
import math
```

```python
x = np.linspace(0, 10, 1000)
P = norm.pdf(x, loc=3, scale=0.5) + norm.pdf(x, loc=7, scale=0.5)

with plt.style.context('seaborn'):
  plt.figure(figsize=(10, 5.5))
  plt.plot(x, P)
```

```python
def kullback_leibler_divergence(p, q):
    return np.sum(p * np.log(p / q))
```

```python
Q_list = [norm.pdf(x, loc=3, scale=0.5),
          norm.pdf(x, loc=5, scale=1.5),
          norm.pdf(x, loc=7, scale=0.5)]
```

## D(P, Q)

```python
names = ['N(3, 0.5)', 'N(5, 1.5)', 'N(7, 0.5)']
for i, Q in enumerate(Q_list):
    print(f"KL(p, {names[i]}): {kullback_leibler_divergence(P, Q):.4f}")
```

```python
with plt.style.context('seaborn'):
  plt.figure(figsize=(10, 5.5))
  plt.plot(x, P, label="p")
  plt.plot(x, Q_list[1], label=names[1], linestyle='dashed')
  plt.legend()
```

## D(Q, P)

```python
for i, Q in enumerate(Q_list):
    print(f"KL(p, {names[i]}): {kullback_leibler_divergence(Q, P):.4f}")
```

```python
with plt.style.context('seaborn'):
  plt.figure(figsize=(10, 5.5))
  plt.plot(x, P, label="p")
  plt.plot(x, Q_list[0], label=names[0], linestyle='dashed', linewidth=5)
  plt.plot(x, Q_list[2], label=names[2], linestyle='dashed', linewidth=5)
  plt.legend()
```
